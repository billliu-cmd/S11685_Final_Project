```python
import pandas as pd
import numpy as np
```

```python
def make_time_splits(
    panel: pd.DataFrame,
    train_frac: float = 0.70,
    val_frac: float = 0.15,
):
    dates = pd.Index(sorted(panel["date"].unique()))
    n = len(dates)

    train_end = int(n * train_frac)
    val_end = int(n * (train_frac + val_frac))

    train_dates = pd.DatetimeIndex(dates[:train_end])
    val_dates = pd.DatetimeIndex(dates[train_end:val_end])
    test_dates = pd.DatetimeIndex(dates[val_end:])

    return train_dates, val_dates, test_dates
```
